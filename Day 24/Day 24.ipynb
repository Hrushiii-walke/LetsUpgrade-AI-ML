# Decision Tree

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn import preprocessing
```

```python
decision_tree=pd.read_excel("Bank_Personal_Loan_Modelling.xlsx",sheet_name=1)
```

```python
decision_tree.head()
```

```python
decision_tree.isnull().sum()
```

There is no any missing value in data

# creating model

```python
tree_model=tree.DecisionTreeClassifier(max_depth=8)
```

```python
decision_tree.columns
```

```python
independent_variable=pd.DataFrame([decision_tree["Age"],decision_tree["Experience"],decision_tree["Income"],decision_tree["Family"],decision_tree["CCAvg"],decision_tree["Education"],decision_tree["Mortgage"],decision_tree["Securities Account"],decision_tree["CD Account"],decision_tree["Online"],decision_tree["CreditCard"]]).T
```

```python
tree_model.fit(X=pd.DataFrame(independent_variable),y=decision_tree["Personal Loan"])
```

```python
with open ("tree.dot",'w') as f:
    f=tree.export_graphviz(tree_model,feature_names=["Income","Age","Experience","Family","CCAvg","Education","Mortgage","Securities Account","CD Account","Online","CreditCard"],out_file=f)
```
